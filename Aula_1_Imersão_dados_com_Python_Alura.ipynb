# Traduzindo para o PT-BR as colunas do DataFrame

```python
# Nomes das colunas

df.columns
```

```python
columns_translation = {
    'work_year': 'ano_trabalho',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'titulo_cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_funcionario',
    'remote_ratio': 'proporcao_remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'tamanho_empresa'
}

''' Renomeando as colunas para o PT-BR '''

df = df.rename(columns=columns_translation)
print(df.columns)
```

# Soma dos agrupamentos de cada coluna

```python
# Agrupa a coluna e soma os resultados

df["nivel_experiencia"].value_counts()
```

```python
df["tipo_emprego"].value_counts()
```

```python
df["proporcao_remoto"].value_counts()
```

```python
nivel_experiencia_translation = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

df['nivel_experiencia'] = df['nivel_experiencia'].replace(nivel_experiencia_translation)

df['nivel_experiencia'].value_counts()
```

```python
tipo_emprego_translation = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Autônomo'
}

df['tipo_emprego'] = df['tipo_emprego'].replace(tipo_emprego_translation)

df['tipo_emprego'].value_counts()
```

```python
tamanho_empresa_translation = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa_translation)

df['tamanho_empresa'].value_counts()
```

```python
proporcao_remoto_translation = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Totalmente Remoto'
}

df['proporcao_remoto'] = df['proporcao_remoto'].replace(proporcao_remoto_translation)

df['proporcao_remoto'].value_counts()
```

```python
# Completando salários nulos com média e mediana

import numpy as np

''' np.nam -> Valores nulos '''

df_salarios = pd.DataFrame({
    "nome" : ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    "salario" : [4000, np.nan, 5000, np.nan, 100000]
})


''' Calcula mediana e substitui os nulos pela média '''

df_salarios["salario_media"] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

''' Calcula mediana e substitui os nulos pela mediana '''

df_salarios["salario_mediana"] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
# Preeenchendo temperaturas nulas com ffill -> Completa com o valor anterior

df_temperaturas = pd.DataFrame({
    "dia" : ["Segunda-feira", "Terça-feira", "Quarta-feira", "Quinta-feira", "Sexta-feira"],
    "temperatura" : [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["temperatura"].ffill()
df_temperaturas
```

```python
# Preeenchendo temperaturas nulas com bfill -> Completa com o valor posterior

df_temperaturas = pd.DataFrame({
    "dia" : ["Segunda-feira", "Terça-feira", "Quarta-feira", "Quinta-feira", "Sexta-feira"],
    "temperatura" : [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_bfill"] = df_temperaturas["temperatura"].bfill()
df_temperaturas
```

```python
# Preenchendo uma nova coluna com "Não informado" onde há valores nulos

df_cidades = pd.DataFrame({
    "nome" : ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    "cidade" : ["São Paulo", np.nan, "Curitiba", np.nan, "Belém"]
})

df_cidades['cidade_preenchida'] = df_cidades["cidade"].fillna("Não informado")
df_cidades
```

# Media Salarial por Nível de Senioridade

```python
#Utilizando o plotly

import plotly.express as px

# Calculate the average salary per experience level
df_salario_por_senioridade = df_limpo.groupby('nivel_experiencia')['salario_em_usd'].mean().reset_index()

# Define the order for seniority levels
ordem_senioridade = ['Júnior', 'Pleno', 'Sênior', 'Executivo']

# Create the bar chart using Plotly Express
fig = px.bar(
    df_salario_por_senioridade,
    x='nivel_experiencia',
    y='salario_em_usd',
    title='Média Salarial por Nível de Senioridade (USD)',
    labels={
        'nivel_experiencia': 'Nível de Experiência',
        'salario_em_usd': 'Média Salarial Anual (USD)'
    },
    category_orders={'nivel_experiencia': ordem_senioridade},
    color='nivel_experiencia', # Color bars by seniority level
    color_discrete_sequence=px.colors.qualitative.Set2 # Choose a color sequence
)

# Update layout for better readability
fig.update_layout(xaxis_title='Nível de Experiência', yaxis_title='Média Salarial Anual (USD)')

# Show the plot
fig.show()
```

```python

# Gráfico Pizza e Rosca

remoto_contagem = df_limpo['proporcao_remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_emprego', 'proporcao_remoto']
fig = px.pie(
    remoto_contagem,
    names='tipo_emprego',
    values='proporcao_remoto',
    title='Proporção dos Tipos de Trabalho Remoto',
    hole=0.5
)

fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
import plotly.express as px

# Calcula a média de salário por localização da empresa
df_salario_por_pais = df_limpo.groupby('localizacao_empresa')['salario_em_usd'].mean().reset_index()

# Classifica os dados pela média salarial em ordem decrescente para melhor visualização
df_salario_por_pais = df_salario_por_pais.sort_values(by='salario_em_usd', ascending=False)

# Cria o gráfico de barras usando Plotly Express
fig = px.bar(
    df_salario_por_pais.head(20), # Exibe os 20 principais países para melhor legibilidade
    x='localizacao_empresa',
    y='salario_em_usd',
    title='Média Salarial por Localização da Empresa (Top 20)',
    labels={
        'localizacao_empresa': 'Localização da Empresa',
        'salario_em_usd': 'Média Salarial Anual (USD)'
    },
    color='localizacao_empresa', # Colore as barras por país
    color_discrete_sequence=px.colors.qualitative.Plotly # Escolhe uma sequência de cores
)

# Atualiza o layout para melhor legibilidade
fig.update_layout(xaxis_title='País', yaxis_title='Média Salarial Anual (USD)')

# Exibe o gráfico
fig.show()
```
